```python
#to create plots inside the notebook
%matplotlib inline
from pylab import *
import matplotlib.pyplot as plt
import numpy
```

```python
x= numpy.linspace(-5,5, 1000)
y= x**3
```

```python
fig= plt.figure(figsize=(10, 5), dpi=100) 
ax= fig.add_subplot(121)
ax.plot(x,y, 'r')
ax1= fig.add_axes([0.15, 0.65, 0.2, 0.2]) # left, bottom, width, height
ax1.plot(y,x, 'b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax= fig.add_subplot(122)
ax.plot(x, y**2)
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.tight_layout() #adjust the positions of labels to not overlap
```

```python
#matplotlib.rcParams.update({'font.size': 18, 
#                            'font.family': 'serif'})

matplotlib.rcParams.update({'font.size': 18, 
                            'font.family': 'STIXGeneral', 
                            'mathtext.fontset': 'stix'})

fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$") # r is for raw text
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('title')
```

```python
# scientific notation
matplotlib.rcParams.update({'font.size': 18, 
                            'font.family': 'serif'})
fig = plt.figure()
ax= fig.add_subplot(111)      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
```
